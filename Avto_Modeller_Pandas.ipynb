```python
import pandas as pd
```

```python
data = {
    'marka': ['BMW','Honda','Toyota','Nissan','Lambo','Ford'],
    'model': ['Alpina','Insight','Avalon','Sunny','Aventador','Fusion'],
    'qiymet': [25000,38000,45000,18000,68000,28000]
}
df = pd.DataFrame(data)
```

```python
min_row = df[df['qiymet'] == df['qiymet'].min()] #Minimum və maksimum qiymətlər
max_row = df[df['qiymet'] == df['qiymet'].max()]

print("Minimum qiymət\n", min_row)
print("Maksimum qiymət:\n", max_row)
```

```python
ortalama = df['qiymet'].mean()   #Qiymətlərin ortalaması
print("Qiymətlərin ortalaması:", ortalama)
```

```python
siralama_df = df.sort_values(by='qiymet', ascending=True)  #Qiymətlər artan sıra ilə
print(siralama_df)
```

```python
boyuk_25000 = df[df['qiymet'] > 25000] #Qiymeti 25000-dən böyük olan avtomobillər
print("25000-dən böyük qiyməti olanlar", boyuk_25000)
```

```python
df['+20% artım'] = df['qiymet'] * 1.20  #Qiymətləri 20% artırın
print(df)
```

```python
df['-10% endirim'] = df['qiymet'] * 0.90   #Qiymətlərə 10% endirim
print(df)
```

```python
import matplotlib.pyplot as plt   #"Maşın modelleri üzrə orijinal və endirimli qiymətlər"

plt.bar(df['model'], df['qiymet'], label='Orijinal Qiymət', alpha=0.7, color='red')
plt.bar(df['model'], df['-10% endirim'], label='10% endirim', alpha=0.7, color='blue')

plt.title("Maşın modelleri üzrə orijinal və endirimli qiymətlər")
plt.xlabel("Model")
plt.ylabel("Qiymət (AZN)")
plt.xticks()
plt.legend()
plt.show()
```

```python
import pandas as pd         #Avtomobillərin öz qiymətləri və 20% artım qiymətləri
import matplotlib.pyplot as plt
import numpy as np

df = pd.DataFrame({
    'marka': ['BMW', 'Honda', 'Toyota', 'Nissan', 'Lambo', 'Ford'],
    'model': ['Alpina', 'Insight', 'Avalon', 'Sunny', 'Aventador', 'Fusion'],
    'qiymet': [25000, 38000, 45000, 18000, 68000, 28000],
    '+20% artım': [30000, 45600, 54000, 21600, 81600, 33600]
})

x = np.arange(len(df['model']))
width = 0.35

plt.bar(x - width/2, df['qiymet'], width, label='Orijinal Qiymət', color='red', alpha=0.7)
plt.bar(x + width/2, df['+20% artım'], width, label='+20% Artım', color='blue', alpha=0.7)

plt.xticks(x, df['model'], rotation=45)
plt.xlabel("Model")
plt.ylabel("Qiymət (AZN)")
plt.title("Orijinal qiymətləri və 20% artım")
plt.legend()
plt.show()
```
